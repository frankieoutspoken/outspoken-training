### Model Selection — Choosing the Right Brain for the Job

Not all LLMs are the same. They come in different sizes — think of it like hiring: you don't need a senior architect to answer a yes/no question.

> **Vocab: Model** — The trained AI system. Different models = different tradeoffs between speed, cost, and intelligence.

> **Vocab: Latency** — How long it takes to get a response back. Smaller models are faster.

> **Vocab: Reasoning** — The model's ability to think through complex, multi-step problems. Bigger models reason better.

| Model | Speed | Cost | Best for |
|---|---|---|---|
| **Haiku** | Fastest | Cheapest | Simple tasks — classify, extract, quick answers |
| **Sonnet** | Balanced | Mid-range | Most everyday work — writing, analysis, coding |
| **Opus** | Slower | Most expensive | Complex reasoning, nuanced tasks, long documents |

# Part 3: Build an Agent — Midnight Cosmetics

## The Scenario

**Midnight Cosmetics** wants to create a customer service agent for their website. They've hired you — a solutions architect — to design and build it.

## Step 1: Understand the Goals

What should this agent do?

1. **Answer customer queries** — about products, returns, policies, general info
2. **Handle returns** — look up orders, check the return window, guide the customer
3. **Generate personalized beauty profiles** — create a branded document from quiz results
4. **Escalate to a human** — when it can't help, collect the customer's email and hand off

### What Does "Good" Look Like?

Before we build, we need to define success. Here are a few scenarios and what a *good* agent response looks like:

**Scenario: Return request**
- Customer says: *"I want to return my order"*
- Agent should: Ask for the order number, name, and email
- Agent looks up the order → checks if it's within the 7-day return window → checks if it's a sale item
- If eligible: provides return instructions
- If not eligible: explains why and offers alternatives
- If customer disputes: escalates to a human

**Scenario: Product question**
- Customer says: *"What do you have for oily skin?"*
- Agent should: Search the product catalog and recommend relevant products (Stardust Setting Powder, etc.)
- Should NOT hallucinate products that don't exist

**Scenario: Beauty profile**
- Customer says: *"I took the quiz — oily skin, natural look, acne concerns"*
- Agent should: Generate a personalized, branded document with product recommendations tailored to their profile

## Step 2: What Do We Need?

Think about this agent like a **person** you're hiring for this job. What would that person need to do their work?

**They need a brain** — that's the LLM. A large language model is the "brain" of the agent. It reasons, understands language, and generates responses.

But a brain alone isn't enough. To actually *work*, this person would need:

- **Access to the product catalog** — they'd need to walk over to the filing cabinet and look up products. That's a *tool*.
- **The return policy** — they'd need to know the rules. That's *instructions*.
- **The order database** — they'd need to look up a specific customer's order. That's another *tool*.
- **The ability to create documents** — they'd need to write up a beauty profile. That's a *skill*.
- **The ability to call for help** — when something's beyond them, they hand off to a manager. That's *escalation*.

And before any of that works, they need **a key to get into the building** — that's the API key. It's your credential to access the LLM.

> **Vocab: API Key** — A secret credential that gives you access to a model. Like a key to a building. Without it, you can't connect. You get one from Anthropic (for Claude) or OpenAI (for embeddings).

Now — would **one person** handle all of this? Product questions, returns, beauty profiles, AND escalation? Probably not. You'd hire specialists. That's **multi-agent architecture** — multiple specialized agents, each with their own tools, coordinated by a router that decides who handles what.

> **Vocab: Multi-agent** — Multiple specialized agents working together, coordinated by a router.

> **Vocab: Handoff** — When one agent passes the conversation to another. The customer doesn't see it happen.

## Step 3: Map the Architecture

![Midnight Cosmetics Agent Architecture](assets/midnight_architecture.png)

> **Vocab: Orchestration** — The router agent is the orchestrator. It decides which specialist handles each request.

> **Vocab: Autonomy** — Each agent has a different level. The Product FAQ agent just answers questions. The Returns agent checks rules and may escalate. The Beauty Profile agent creates documents.

> **Vocab: Guardrails** — Constraints on what an agent can do. The returns agent can't override the return policy. The product agent can't make up products that don't exist.

## Step 4: Build the Prerequisites

We need to prepare the data and tools before we build the agent. The files are already in the `data/` folder:

- `data/midnight_products.csv` — Product catalog (10 products with names, categories, prices, descriptions)
- `data/midnight_orders.csv` — Order database (8 orders with dates, statuses, sale flags)
- `data/midnight_return_policy.md` — Return policy with rules and escalation criteria
- `data/midnight_knowledge_base/` — Beauty knowledge base (skincare routines, ingredient guides, application tips, FAQs)

### Which Data Goes Where?

Not everything gets treated the same way. The right approach depends on the type of data:

| Data | Type | Approach | Why |
|---|---|---|---|
| **Product catalog** | Structured (CSV) | **Tool** — CSV lookup | It has clear columns. You can filter by category, search by name, look up prices. A database query is simpler and more reliable. |
| **Return policy** | Short, strict rules | **Direct instructions** | Must be followed exactly every time. The agent sees the full policy on every request. No risk of retrieving the wrong chunk. |
| **Order database** | Structured (CSV) | **Tool** — CSV lookup | Structured data with order IDs, dates, statuses. Look up by order ID. |
| **Knowledge base** | Unstructured articles | **RAG — Vector Store** | Long-form text about skincare, ingredients, application tips. A customer asks "how do I avoid cakey foundation?" — the agent needs to search by *meaning*, not keywords. |

### The Knowledge Base → Vector Store (RAG)

This is where RAG shines. The knowledge base has articles about skincare routines, ingredient guides, and application tips. When a customer asks a question like *"what's good for blemish-prone skin?"* the agent needs to find the right article even if the word "blemish" doesn't appear — maybe the article talks about "acne-prone" or "breakouts" instead. That's semantic search.

Here's how it works under the hood:

![How RAG Works](assets/rag_architecture.png)

**The Indexing Pipeline** (happens once, when you upload):
1. **Raw documents** — your knowledge base articles (markdown files)
2. **Chunking** — split into smaller, searchable pieces. A long article about skincare routines might become 4-5 chunks. This matters because smaller chunks are more precise — you retrieve just the relevant paragraph, not the whole article.
3. **Embedding** — each chunk gets converted into a vector (a list of numbers that capture its meaning). The text gets tokenized and transformed into a numerical representation that captures what it's *about*.
4. **Vector Store** — the embeddings get stored in a searchable database

**The Query Pipeline** (happens every time a customer asks):
1. The customer's question gets **embedded** into a vector
2. The vector store **searches** for chunks with the most similar vectors — this is why "blemish-prone" finds "acne-prone" content
3. Those relevant chunks get passed to the **LLM** along with the question
4. The LLM generates an answer **grounded in the actual knowledge base**

> **Vocab: Chunking** — Breaking documents into smaller pieces for better search accuracy. A 10-page doc might become 50 chunks. Chunk size matters — too big and you retrieve irrelevant info, too small and you lose context.

> **Vocab: RAG (Retrieval-Augmented Generation)** — The whole pipeline: retrieve relevant data, then generate a response using that data as context. The model doesn't memorize your docs — it searches them on demand.

## Step 5: Build It — In Claude Code

Now we leave the notebook. Open your terminal, navigate to this project, and launch **Claude Code**.

### What is Markdown?

The build prompt for our agent is written in a `.md` file — that's **markdown**. Markdown is a simple formatting language:
- `#` = heading, `##` = subheading
- `**bold**`, `*italic*`
- `-` for bullet points
- ``` for code blocks
- `|` for tables

It's how most technical documentation is written, and it's what Claude Code reads natively. You'll see it everywhere — GitHub READMEs, project docs, agent instructions.

### The Build Process

We've prepared two files in `midnight-cosmetics-agent/`:

- **`CLAUDE.md`** — Project structure and conventions. Claude Code reads this automatically to understand the project.
- **`AGENT_PROMPT.md`** — The full build prompt with every detail: agent architecture, model selection, tools, guardrails, parameters, session tracking, handoff logic, and test scenarios.

Open the prompt file and read through it — every decision is documented:
- Which model each agent uses and why (Haiku for routing, Sonnet for reasoning)
- What tools each agent has access to
- The exact return policy rules baked into the returns agent
- Session IDs that track every interaction across agent handoffs
- Guardrails for each agent

### Build Steps

1. Open Claude Code in `midnight-cosmetics-agent/`
2. Type `/new-sdk-app` — this is the **skill** for scaffolding Claude Agent SDK projects. It sets up the project structure, dependencies, and configuration.
3. Choose **Python** when prompted
4. Then paste: `"Read AGENT_PROMPT.md and build the full multi-agent system described there."`
5. Claude will read the prompt, build each agent, create the tools, wire up the Streamlit frontend, and set up session tracking.

> **Vocab: Skill** — A slash command that gives Claude Code specialized knowledge for a task. `/new-sdk-app` knows the Claude Agent SDK patterns. You don't have to explain the SDK — the skill already knows it.

> **Vocab: Session ID** — A unique identifier (UUID) for each conversation. It gets passed to every agent so you can trace the full conversation path: which agents handled what, what tools were called, what decisions were made. Essential for debugging and auditing.

## Vocab Glossary

| Term | Definition |
|---|---|
| **LLM** | Large Language Model — generates text by predicting the next token |
| **Token** | A chunk of text (~3/4 of a word). The unit models think in |
| **Model** | The trained AI system. Different models = speed/cost/quality tradeoffs |
| **Latency** | How long it takes to get a response |
| **Reasoning** | A model's ability to work through multi-step problems |
| **Context Window** | Total text a model can see at once (prompt + docs + response) |
| **System Prompt** | Hidden instructions that shape the model's behavior |
| **Temperature** | Controls randomness — low = focused, high = creative |
| **Hallucination** | When the model confidently says something untrue |
| **Zero-shot** | Performing a task with no examples, just an instruction |
| **Few-shot** | Providing examples before asking the model to perform |
| **Chain of Thought** | Asking the model to reason step by step |
| **API** | A way to interact with the model via code |
| **API Key** | Secret credential that gives you access to a model |
| **Tool** | A capability beyond text generation (search, APIs, code) |
| **Integration / Connector** | Pre-built tool connections in products like Claude |
| **RAG** | Retrieval-Augmented Generation — grounding answers in your data |
| **Chunking** | Breaking documents into smaller searchable pieces |
| **Embeddings** | Converting text to numbers for similarity search |
| **Vector Store** | A database for storing and searching embeddings |
| **Agent** | AI that perceives, decides, and acts toward a goal |
| **Guardrails** | Rules constraining what an agent can and cannot do |
| **Human-in-the-loop** | Agent asks a human before acting |
| **Autonomy** | How much an agent can do without asking permission |
| **MCP** | Model Context Protocol — universal standard for connecting AI to tools |
| **Skill** | Reusable instructions that specialize an agent for a job |
| **Handoff** | One agent passing a task to another |
| **Orchestration** | Managing multiple agents working together |
| **Multi-agent** | System of specialized agents coordinated by a router |
